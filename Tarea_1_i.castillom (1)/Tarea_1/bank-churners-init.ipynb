```python
# número de clientes perdido y no perdidosn
counts = df.Attrition_Flag.value_counts()
perc_churn = (counts.iloc[1] / (counts.iloc[0] + counts.iloc[1])) * 100
print(f'Churn Rate = {perc_churn:.1f}%')
```

La función calcula la tasa de abandono como el porcentaje de `Attrited Customer` sobre el total de clientes (la suma de `Existing Customer` y `Attrited Customer`), multiplicado por 100. Utiliza `iloc` para seleccionar valores específicos en el DataFrame `counts`(almacena el conteo por categoría de la columna `Attrition_Flag` utilizando la función `value_counts`), donde `Existing Customer` está en la fila 0 y `Attrited Customer` en la fila 1. Finalmente, imprime el valor calculado en `perc_churn` junto con el texto "Churn Rate", formateado como porcentaje.

## Pre-Processing

```python
# Conversión de características
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Attrited Customer': 1, 'Existing Customer': 0})
```

Esta función convierte los valores categóricos de las columnas `Gender` y `Attrition_Flag` en valores numéricos (1 y 0) en el DataFrame `df`.

```python
catcols = df.select_dtypes(exclude = ['int64','float64']).columns
intcols = df.select_dtypes(include = ['int64']).columns
floatcols = df.select_dtypes(include = ['float64']).columns

# codificación 
df = pd.get_dummies(df, columns = catcols)
   
print('New Number of Features: %d'%(df.shape[1]))  
```

Esta función identifica las columnas categóricas, enteras y de punto flotante en `df`, aplica codificación one-hot a las categóricas, y luego imprime el nuevo número de columnas del DataFrame.
