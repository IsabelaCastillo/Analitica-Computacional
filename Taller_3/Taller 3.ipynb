## Evaluar el Modelo

```python
y_pred = linreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
import numpy as np

# mean absolute error
MAE = metrics.mean_absolute_error(y_test, y_pred)

# mean squared error
MSE = metrics.mean_squared_error(y_test, y_pred)

# root mean squared error
RMSE = np.sqrt(MSE)
```

```python
featuresb = ['X1 transaction date','X2 house age',
                           'X3 distance to the nearest MRT station','X4 number of convenience stores'
                           ,'X5 latitude','X6 longitude']

X = data[featuresb]
y = data['Y house price of unit area']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

linreg.fit(X_train, y_train)

y_pred = linreg.predict(X_test)

MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)

print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
```

### Validacin Cruzada

```python
from sklearn.model_selection import cross_val_score

# usar MSE - error cuadrtico medio
scores = cross_val_score(linreg, X, y, cv=5, scoring='neg_mean_squared_error')
mse_scores = - scores
print(mse_scores)
```

```python
# calcular RMSE
rmse_scores = np.sqrt(mse_scores)
print(rmse_scores)
```

```python
# RMSE promedio a travs de todos los grupos
print(rmse_scores.mean())
print(mse_scores.mean())
```

### Statsmodel

```python
import statsmodels.api as sm

X = data[features]
y = data['Y house price of unit area']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# agregar constante explticamente
X_train = sm.add_constant(X_train)

# regresin usando mnimos cuadrados ordinarios (ordinary least squares - OLS) 
model = sm.OLS(y_train, X_train).fit()

# resumen de resultados
print(model.summary())
```

### Determina la influencia de las Observaciones

```python
fig = sm.graphics.influence_plot(model, criterion="cooks")
```

```python
# disntacia de Cook
model_cooksd = model.get_influence().cooks_distance[0]

# get length of df to obtain n
n = X_train.shape[0]

# umbral
critical_d = 4/n
print('Umbral con distancia de Cook:', critical_d)

# puntos que podran ser ourliers con alta influencia
out_d = model_cooksd > critical_d

print(X_train.index[out_d], "\n", model_cooksd[out_d])
```
